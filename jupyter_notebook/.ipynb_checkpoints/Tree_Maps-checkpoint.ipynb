```python
def read_shapefile(sf):
    fields = [x[0] for x in sf.fields][1:]
    records = sf.records()
    shapes = [s.points for s in sf.shapes()]

    df = pd.DataFrame(columns=fields, data=records)
    df = df.assign(coords=shapes)

    return df
```

```python
def plot_shape(id, s=None):
    """ PLOTS A SINGLE SHAPE """
    plt.figure()
    ax = plt.axes()
    ax.set_aspect('equal')
    shape_ex = sf.shape(id)
    x_lon = np.zeros((len(shape_ex.points),1))
    y_lat = np.zeros((len(shape_ex.points),1))
    for ip in range(len(shape_ex.points)):
        x_lon[ip] = shape_ex.points[ip][0]
        y_lat[ip] = shape_ex.points[ip][1]

    plt.plot(x_lon,y_lat) 
    x0 = np.mean(x_lon)
    y0 = np.mean(y_lat)
    plt.text(x0, y0, s, fontsize=10)
    # use bbox (bounding box) to set plot limits
    plt.xlim(shape_ex.bbox[0],shape_ex.bbox[2])
    return x0, y0
```

```python
# Plot neighborhood
neighborhood = 'Japantown'
dist_id = df[df.name == neighborhood].index.get_values()[0]
plot_shape(dist_id, neighborhood)
```

```python
def plot_map(sf, x_lim = None, y_lim = None, figsize = (11,9)):
    '''
    Plot map withlim coordinates
    '''
    plt.figure(figsize = figsize)
    id=0
    for shape in sf.shapeRecords():
        for i in range(len(shape.shape.parts)):
            i_start = shape.shape.parts[i]
            if i==len(shape.shape.parts)-1:
                i_end = len(shape.shape.points)
            else:
                i_end = shape.shape.parts[i+1]
            x = [i[0] for i in shape.shape.points[i_start:i_end]]
            y = [i[1] for i in shape.shape.points[i_start:i_end]]
            plt.plot(x, y, 'k')
        
        if (x_lim == None) & (y_lim == None):
            x0 = np.mean(x)
            y0 = np.mean(y)
            plt.text(x0, y0, id, fontsize=10)
        id = id+1
    
    if (x_lim != None) & (y_lim != None):     
        plt.xlim(x_lim)
        plt.ylim(y_lim) 
```

```python
plot_map(sf)
```

```python
def plot_map_fill_multiple_ids(title, districts, sf, 
                                               x_lim = None, 
                                               y_lim = None, 
                                               figsize = (11,9), 
                                               color = 'r'):
    '''
    Plot map with lim coordinates
    '''
    plt.figure(figsize = figsize)
    fig, ax = plt.subplots(figsize = figsize)
    fig.suptitle(title, fontsize=16)

    for shape in sf.shapeRecords():
        for i in range(len(shape.shape.parts)):
            i_start = shape.shape.parts[i]
            if i==len(shape.shape.parts)-1:
                i_end = len(shape.shape.points)
            else:
                i_end = shape.shape.parts[i+1]
            x = [i[0] for i in shape.shape.points[i_start:i_end]]
            y = [i[1] for i in shape.shape.points[i_start:i_end]]
            plt.plot(x, y, 'k')
            
    for id in districts:
        shape_ex = sf.shape(id)
        x_lon = np.zeros((len(shape_ex.points),1))
        y_lat = np.zeros((len(shape_ex.points),1))
        for ip in range(len(shape_ex.points)):
            x_lon[ip] = shape_ex.points[ip][0]
            y_lat[ip] = shape_ex.points[ip][1]
        ax.fill(x_lon,y_lat, color)
             
        x0 = np.mean(x_lon)
        y0 = np.mean(y_lat)
        plt.text(x0, y0, id, fontsize=10)
    
    if (x_lim != None) & (y_lim != None):     
        plt.xlim(x_lim)
        plt.ylim(y_lim)
        
    #fig.savefig(title)
```

```python
def plot_neighborhoods(title, neighborhoods, color):
    neighborhood_id = []
    for i in neighborhoods:
        try:
            neighborhood_id.append(df[df.name == i]
                             .index.get_values()[0])
        except:
            pass
    plot_map_fill_multiple_ids(title, neighborhood_id, 
                                       sf, 
                                       x_lim = None, 
                                       y_lim = None, 
                                       figsize = (11,9), 
                                       color = color); 
```

```python
title = 'Popular spots'
neighborhoods = ['Mission', 'Inner Sunset', 'Cole Valley', 'Marina']
color = '#00FF0030'
plot_neighborhoods(title, neighborhoods, color ) 
```

```python
def calc_color(data, color=None):
    if color   == 1: color_sq = ['#dadaebFF','#bcbddcF0','#9e9ac8F0','#807dbaF0','#6a51a3F0','#54278fF0']; colors = 'Purples';
    elif color == 2: color_sq = ['#c7e9b4','#7fcdbb','#41b6c4','#1d91c0','#225ea8','#253494']; colors = 'YlGnBu';
    elif color == 3: color_sq = ['#f7f7f7','#d9d9d9','#bdbdbd','#969696','#636363','#252525']; colors = 'Greys';
    elif color == 4: color_sq = ['#c0ec83', '#a7e074', '#96cc39', '#6ba32d', '#547a1d', '#3b5c0a']; colors = 'Greens';
    elif color == 9: color_sq = ['#ff0000','#ff0000','#ff0000','#ff0000','#ff0000','#ff0000']
    else:            color_sq = ['#ffffd4','#fee391','#fec44f','#fe9929','#d95f0e','#993404']; colors = 'YlOrBr';
    new_data, bins = pd.qcut(data, 6, retbins=True, labels=list(range(6)))
    color_ton = []
    for val in new_data:
        color_ton.append(color_sq[val]) 
    if color != 9:
        colors = sns.color_palette(colors, n_colors=6)
        sns.palplot(colors, 0.6);
        for i in range(6):
            print ("\n"+str(i+1)+': '+str(int(bins[i]))+" => "+str(int(bins[i+1])-1))
    return color_ton, bins;
```

```python
def plot_map_fill_multiple_ids_tone(sf, title, districts, dist_names, dist_color,
                                     print_id, color_ton, 
                                     bins, 
                                     x_lim = None, 
                                     y_lim = None, 
                                     figsize = (44,36)):
        
    plt.figure(figsize = figsize)
    fig, ax = plt.subplots(figsize = figsize)
    fig.suptitle(title, fontsize=16)

    for shape in sf.shapeRecords():
        for i in range(len(shape.shape.parts)):
            i_start = shape.shape.parts[i]
            if i==len(shape.shape.parts)-1:
                i_end = len(shape.shape.points)
            else:
                i_end = shape.shape.parts[i+1]
            x = [i[0] for i in shape.shape.points[i_start:i_end]]
            y = [i[1] for i in shape.shape.points[i_start:i_end]]
            plt.plot(x, y, 'k')
            
    for id in districts:
        shape_ex = sf.shape(id)
        x_lon = np.zeros((len(shape_ex.points),1))
        y_lat = np.zeros((len(shape_ex.points),1))
        for ip in range(len(shape_ex.points)):
            x_lon[ip] = shape_ex.points[ip][0]
            y_lat[ip] = shape_ex.points[ip][1]
        ax.fill(x_lon, y_lat, dist_color[id])    
        #ax.fill(x_lon, y_lat, color_ton[districts.index(id)])
        if print_id != False:
            try:
                x0 = np.mean(x_lon)
                y0 = np.mean(y_lat)
                plt.text(x0, y0, dist_names[id], fontsize=10)
            except:
                print(i, len(dist_names))
    if (x_lim != None) & (y_lim != None):     
        plt.xlim(x_lim)
        plt.ylim(y_lim)
        
    fig.savefig(title);

def plot_trees_data(sf, title, data=None, 
                      color=None, print_id=False):
    
    districts = data.keys()
    color_ton, bins = calc_color(list(data.values()), color)
    df = read_shapefile(sf)
    dist_id = []
    dist_names = {}
    dist_color = {}
    j = 0
    for dist, value in data.items():
        try:
            id_fld = df[df.name == dist].index.get_values()[0]
            dist_id.append(id_fld)
            dist_names[id_fld] = dist
            dist_color[id_fld] = color_ton[j]
        except:
            print(dist, " not found")
        j += 1
    
    plot_map_fill_multiple_ids_tone(sf, title, dist_id, dist_names, dist_color,
                                     print_id, 
                                     color_ton, 
                                     bins, 
                                     x_lim = None, 
                                     y_lim = None, 
                                     figsize = (33,23));
```

# Average number of trees in neighborhood (dependent on neighborhood size)

```python
sf.shapeType
```

```python
all_shapes = sf.shapes()
```

```python
i = 0
area_dict = {}
sf.__dict__
for shape in all_shapes:
    rec = sf.record(i)
    area_dict[rec[1]] = rec[2]
#     print 'attributes: %s' % rec
#     print 'bbox: %s' % shape.bbox
    area = rec[2]
    i += 1

# print(len(area_dict))
# print(area_dict)
```

```python
def divide_dict(dividends, divisors):
    ret = dict()
    for key, dividend in dividends.items():
        ret[key] = dividend/(divisors.get(key, 1)*1.0)*1000000
    return ret

scaled_agg_trees_dict = divide_dict(agg_trees_dict, area_dict)
sorted_scaled_agg_trees_dict = sorted(scaled_agg_trees_dict.items(), key=lambda x: x[1], reverse=True)
for k, v in sorted_scaled_agg_trees_dict:
    print(k, v)
```

```python
title = 'Scaled Tree Distrubution in SF Region'
plot_trees_data(sf, title, data = scaled_agg_trees_dict, color = 4, print_id = True)
```

```python
x = [sorted_scaled_agg_trees_dict[i][0] for i in range(len(sorted_scaled_agg_trees_dict))]
y = [sorted_scaled_agg_trees_dict[i][1] for i in range(len(sorted_scaled_agg_trees_dict))]
```
